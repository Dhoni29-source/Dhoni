```python
# Menggunakan persamaan fungsi f(x): 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
# Source code lengkap dalam 1 cell
import numpy as np
import matplotlib.pyplot as plt

# Mencari hasil turunan pertama dan kedua menggunakan semua metode dari finite different 
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
x = 0.1
h = 0.01

print('Hasil turunan menggunakan finite different:')
print("\t\t\t f'(x)\t\t f''(x)\t\t ")
# Forward different
dff1 = (f(x+h) - f(x))/h
dff2 = (f(x+2*h) - 2*f(x+h) + f(x))/h**2
print("Forward different\t%f\t%f" % (dff1, dff2))

# Backward different
dff1 = (f(x) - f(x-h))/h
dff2 = (f(x) - 2*f(x-h) + f(x-2*h))/h**2
print("Backward different\t%f\t%f" % (dff1, dff2))

# Central different
dff1 = (f(x+h) - f(x-h))/(2*h)
dff2 = (f(x+h) - 2*f(x) + f(x-h))/h**2
print("Central different\t%f\t%f" % (dff1, dff2))


## Mencari hasil turunan pertama sampai keempat menggunakan central different
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
x = 0.1
h = 0.001

print('\nHasil turunan pertama sampai turunan keempat')
print('Finite different Metode Central :')
print("f'(x)\t\t f''(x)\t\t f'''(x)\tf''''(x) ")
dff1 = (f(x+h) - f(x-h))/(2*h)
dff2 = (f(x+h) - 2*f(x) + f(x-h))/h**2
dff3 = (f(x+2*h) - 2*f(x+h) + 2*f(x-h) - f(x-2*h))/(2*h**3) #Udah bener
dff4 = (f(x+2*h) - 4*f(x+h) + 6*f(x) - 4*f(x-h) + f(x-2*h))/(h**4) #Udah bener
print("%f\t%f\t%f\t%f" % (dff1, dff2, dff3, dff4))


## Plotting grafik dari 2 hasil turunan dgn metode central different 
# Persamaaan
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
h = 0.001
x = np.linspace(-1,1)

# Central different
dff1 = (f(x+h) - f(x-h))/(2*h)
dff2 = (f(x+h) - 2*f(x) + f(x-h))/h**2

# Plot
plt.plot(x,f(x),'-k', x,dff1,'--b', x,dff2,'-.r')
plt.title("Plotting perbandingan nilai f(x), f'(x), f''(x)", size=15)
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.legend(["f(x)","f'(x)","f''(x)"])
plt.grid()
plt.show()


## Plotting Perbandingan nilai Error dari masing2 metode finite different
# Jika h = 0.1
# Persamaan
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
df= lambda x: 0.5*x**4 - 0.6*x**2 + 0.1

h = 0.1
x = np.linspace(-1,1)
plt.plot(x,df(x),'-k')

# Forward different
dff1 = (f(x+h) - f(x))/h
plt.plot(x,dff1,'-b')

# Backward diff
dff1 = (f(x) - f(x-h))/h
plt.plot(x,dff1,'-r')

# Central different
dff1 = (f(x+h) - f(x-h))/(2*h)
plt.plot(x,dff1,'-g')

# Plot
plt.title('Plotting perbandingan nilai error dari metode finite different, jika h=0.1 ')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.legend(["Theory","Forward","Backward","Central"])
plt.grid()
plt.show()


## Plotting Perbandingan nilai Error dari masing2 metode finite different
# Jika h = 0.01
# Persamaan
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
df= lambda x: 0.5*x**4 - 0.6*x**2 + 0.1

h = 0.01
x = np.linspace(-1,1)
plt.plot(x,df(x),'-k')

# Forward different
dff1 = (f(x+h) - f(x))/h
plt.plot(x,dff1,'-b')

# Backward diff
dff1 = (f(x) - f(x-h))/h
plt.plot(x,dff1,'-r')

# Central different
dff1 = (f(x+h) - f(x-h))/(2*h)
plt.plot(x,dff1,'-g')

# Plot
plt.title('Plotting perbandingan nilai error dari metode finite different, jika h=0.01 ')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.legend(["Theory","Forward","Backward","Central"])
plt.grid()
plt.show()
```
