```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/asus/Anaconda3/DSU/Supervised/adult_data.txt', header = None)
data.head(10)
```

```python
columns = ['age','workclass','fnglwt','education','education_num','marital_status','occupation','relationship','race',\
         'sex','capital_gaim','capital_loss','hours_per_week','native_country','target'] 
```

```python
data.columns = columns
print(data.shape)
data.to_csv("C:/Users/asus/Anaconda3/DSU/Supervised/adult.csv", index=-False)
```

```python
data.isnull().sum()
```

```python
data.describe().T
```

```python
data
```

```python
for column in data.columns:
    if data.dtypes[column] == np.object:
        data[column].value_counts().plot(kind="bar", title=column)
    else:
        data[column].hist()
        plt.title(column)
    plt.show()
```

```python
y = data.pop('target')
```

```python
data.groupby('education').nunique()['education_num']
```

```python
data.drop('education', axis = 1, inplace = True)
print(data.shape)
data.head()
```

```python
data = pd.get_dummies(data)
print(data.shape)
data.head()
```

```python
data.info()
```

```python
#Modelling
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.3, random_state=42)
print('jumlah data train : ',len(X_train))
print('jumlah data test : ',len(X_test))
```

```python
# Metode KNN
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
```

```python
prediksi = model.predict(X_test)
print('Accuracy : ', accuracy_score(prediksi, y_test))
print(confusion_matrix(prediksi, y_test))
```

```python
# Metode Decision Tree
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=10, criterion='entropy')
model.fit(X_train,y_train)
```

```python
prediksi = model.predict(X_test)
print('Accuracy : ', accuracy_score(prediksi, y_test))
print(confusion_matrix(prediksi, y_test))
```

```python
# Metode SVM
from sklearn.svm import SVC

model = SVC()
model.fit(X_train,y_train)
```

```python
prediksi = model.predict(X_test)
print('Accuracy : ', accuracy_score(prediksi, y_test))
print(confusion_matrix(prediksi, y_test))
```
